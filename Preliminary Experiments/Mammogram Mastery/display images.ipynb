```python
import os
import random
import matplotlib.pyplot as plt
import cv2

def display_random_images(dataset_path):
    cancer_path = os.path.join(dataset_path, 'Cancer')
    non_cancer_path = os.path.join(dataset_path, 'Non-Cancer')
    
    cancer_images = random.sample(os.listdir(cancer_path), 2)
    non_cancer_images = random.sample(os.listdir(non_cancer_path), 2)
    
    images = [(os.path.join(cancer_path, img), 'Cancer') for img in cancer_images] + \
             [(os.path.join(non_cancer_path, img), 'Normal') for img in non_cancer_images]
    
    fig, axes = plt.subplots(1, 4, figsize=(12, 4))
    for ax, (img_path, label) in zip(axes, images):
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
        ax.imshow(img, cmap='gray')
        ax.set_title(label, fontsize=12, color='red' if label == 'Cancer' else 'green')
        ax.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
display_random_images("Dataset")
```

```python
display_random_images("Test")
```

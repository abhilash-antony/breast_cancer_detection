```python
import pandas as pd
import os
```

```python
non = pd.read_csv('non_cancerous_img.csv')
can = pd.read_csv('cancerous_images.csv')
```

```python
non.info()
```

```python
can.info()
```

```python
# Concatenate the two DataFrames and reset the index
df = pd.concat([non, can], ignore_index=True).reset_index(drop=True)
df.head()
```

```python
df.drop('Unnamed: 0', axis = 1, inplace = True)
df.head()
```

```python
df.head()
```

```python
# Filter the DataFrame to get rows where 'view' is 'AT'
filtered_df = df[df['view'] == 'AT']
filtered_df
```

```python
# Iimage with view == 'AT'
image_paths = []
for i, row in filtered_df.iterrows():
    # Get the image_id from the column
    image_id = row['image_id']
    
    # Construct the file path (assuming images are in DICOM format, e.g., '.dcm')
    image_path = os.path.join('IMG', f"{image_id}.dcm")  # Modify the extension if needed
    image_paths.append(image_path)
image_paths
```

### **CODE TO DLT IMG BELOW**

```python
# Remove  images for which view == 'AT'
for image_path in image_paths:
    # Check if the file exists
    if os.path.exists(image_path):
        os.remove(image_path)  # Delete the image file
        print(f"Deleted: {image_path}")
    else: 
        print(f"File not found: {image_path}")
```

```python
# Delete those rows from the DataFrame
df = df[df['view'] != 'AT']
```
